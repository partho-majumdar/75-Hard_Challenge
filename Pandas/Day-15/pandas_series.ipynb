```python
import numpy as np; # type: ignore
import pandas as pd; # type: ignore
```

### Series from list

```python
# string

country = ['Bd', 'India', 'Nepal', 'Finland']
pd.Series(country) # 1D data structure where i store 1 by 1 in tabular data column is series
```

```python
# integers

runs = pd.Series([90, 23, 54, 11, 48, 87])
runs
```

```python
# Custom index -> index

marks = [87, 98, 79, 91, 75, 29]
sub = ['P', 'S', 'W', 'X', 'J', 'O']

pd.Series(marks, index = sub) # subs became by index
```

```python
# Set own object type -> name

marks = [87, 98, 79, 91, 75]
sub = ['P', 'S', 'W', 'X', 'J']

pd.Series(marks, index = sub)
pd.Series(marks, index = sub, name = "Student marks")
```

### Series from dict

```python
marks = {
    'p' : 83,
    's' : 89,
    'x' : 71,
    'k' : 94
}

pd.Series(marks, name = "your mark")
```

### Series attribute

```python
# Size -> how many values are in series

marks = {
    'p' : 83,
    's' : 89,
    'x' : 71,
    'y' : 71,
    'k' : 94
}

mark_series = pd.Series(marks, name = "your mark")

# size -> total element
mark_series.size

# dtype
mark_series.dtype

# name 
mark_series.name

# is_unique (False if item repeat)
mark_series.is_unique
pd.Series([1, 2, 1, 3, 4, 2, 4, 5]).is_unique

# index (return all index)
mark_series.index

# values
mark_series.values
```

### Series using read_csv

```python
# Squeeze -> convert dataframe to series

subs_data = pd.read_csv('./subs.csv')
subs_data.head(365)
type(pd.read_csv('./subs.csv')) # this is dataframe
type(pd.read_csv('./subs.csv').squeeze()) # this is series
subs = subs_data.squeeze(True)
subs
```

```python
# index_col -> which col became index

kohli_data = pd.read_csv('./kohli_ipl.csv', index_col = "match_no")
vk = kohli_data.squeeze(True)
type(vk)
vk
```

```python
movie_data = pd.read_csv('./bollywood.csv', index_col = 'movie')
movie = movie_data.squeeze(True)
movie
```

### Series Indexing

```python
# Positive indexing 

x = pd.Series([9, 8, 7, 6, 5, 4, 3, 2, 1])
x[0]
x[1:5]
x[1:5:2]

movie[0]
movie['Uri: The Surgical Strike']
movie[-1]
```

### Editing in series

```python
mark_series[1] = 100 # change using index
mark_series['k'] = 100
mark_series

# without using index
mark_series['aa'] = 79
mark_series

# slicing
runs[2:4] = [123, 69]
runs

# Fancy indexing
runs[[0, 3, 5]] = [0, 0, 0]
runs

movie['Aankhen (2002 film)'] = "Coca Cola"
movie
```

### Series with python function

```python
print(len(subs))
print(type(movie))
print(dir(subs)) # all available function
print(sorted(subs)) # output in list
print(max(subs))
print(min(subs))

# type conversion
list(mark_series)
dict(mark_series)

# membership operator (run in index)
'Why Cheat India' in movie # only search in index
'Emraan Hashmi' in movie.values # search in value

for i in movie: # run for values
    print(i)

for i in movie.index: # search in index
    print(i)

# arithmetic operator (Broadcasting)
100 - mark_series
5 + mark_series

# relational operator
vk >= 50
vk[vk >= 50]
```

```python
# Boolean indexing on series

vk[vk >= 50]
vk[vk >= 100]
vk[vk >= 100].size
vk[vk == 0]

subs[subs > 200].size

# find actor who done more than 20 movie
num_of_movie = movie.value_counts()
num_of_movie > 20
num_of_movie[num_of_movie > 20]
```

```python
# drop_duplicates -> remove duplicate value

temp1 = pd.Series([1, 2, 1, 2, 3, 4, 3, 4, 5, 4, 5, 5, 6, 6])
# print(temp1)
temp1.drop_duplicates()
temp1.drop_duplicates(keep = 'last') # delete all except last

movie.drop_duplicates()

# duplicated -> return boolean

temp1.duplicated()
temp1.duplicated().sum() # total duplicate value
```

```python
# isnull -> boolean return

temp = pd.Series([1, 2, np.nan, 4, 5, np.nan, 7, 8, np.nan, 10])
temp.isnull()
print(temp.size) # count with missing value
print(temp.count()) # count without missing value

temp.isnull().sum() # total number of missing value
```

```python
# dropna -> remove all missing value

temp
temp.dropna()
```

```python
# fillna -> replace missing value

temp.fillna(0)
temp.fillna(temp.mean())
```

```python
# isin -> check multiple item

(vk == 49) | (vk == 99)
vk[(vk == 49) | (vk == 99)]

vk.isin([49, 59, 69, 99, 100])
vk[vk.isin([49, 59, 69, 99, 100])]
```
