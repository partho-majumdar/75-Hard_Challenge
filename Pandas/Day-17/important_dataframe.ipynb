```python
# dropna -> in series drop all missing values for columns & if find any NaN value in dataframe drop entire row 

students['Name'].dropna()
students.dropna()
students
students.dropna(how = 'all') # remove only those row that all value are NaN
students.dropna(subset = ['Name']) # remove only Name = NaN row
students.dropna(subset = ['Name', 'package']) 
```

```python
# fillna -> fill missing values

students['Name'].fillna('Unknown')
students['Name'].fillna('Unknown').hasnans

students.fillna(0)

students['package'] = students['package'].fillna(students['package'].mean())
students['Name'] = students['Name'].fillna(method = 'ffill')
# students['Name'] = students['Name'].ffill() # this is rule 02
students
```

```python
# drop_duplicates -> remove duplicate rows

temp = pd.Series([1, 1, 2, 2, 3, 4, 5, 5])
temp.drop_duplicates()

students.duplicated().sum() # how many rows are duplicate
students.drop_duplicates() # by default keep = 'first'
students.drop_duplicates(keep = 'last') 
```

```python
# apply

temp = pd.Series([10, 20, 30, 40])

def sigmoid(value):
    return 1 / 1 + np.exp(-value)

temp.apply(sigmoid)

point_df = pd.DataFrame({
    '1st Point':[(3, 4), (-6, 5), (0, 0), (4, 5)],
    '2nd Point':[(5, 1), (-9, -2), (3, 7), (2, 6)],
})

point_df

def euclidean(row):
    point_A = row['1st Point']
    point_B = row['2nd Point']

    return ((point_A[0] - point_B[0]) ** 2 + (point_A[1] - point_B[1]) ** 2) ** 0.5

point_df['distance'] = point_df.apply(euclidean, axis = 1)
point_df
```
