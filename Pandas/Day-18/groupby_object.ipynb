```python
import numpy as np # type: ignore
import pandas as pd # type: ignore
```

```python
movies = pd.read_csv('./imdb-top-1000.csv')
movies
```

```python
# Groupby -> make group in data on the basis of some column (categorical)

movies
genres = movies.groupby('Genre')

genres.apply(lambda x:x, include_groups = False)
# genres.apply(lambda x:x)
genres.min()
genres.max()
genres.sum()
```

```python
# Find top 3 genres by total earning

movies.groupby('Genre')['Gross'].sum().sort_values(ascending = False).head(3)
```

```python
# find the genre with highest avg IMDB rating

movies.groupby('Genre')['IMDB_Rating']
movies.groupby('Genre')['IMDB_Rating'].mean()
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending = False).head(3)
```

```python
# find director with most popularity

movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(3)
```

```python
# find number of movies done by each actor
# 1. Convert into group
# 2. Select column
# 3. Use aggregate operation

movies['Star1'].value_counts().head(5) # R1
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending = False).head(5) # R2
```

```python
# find total number of group

len(movies.groupby('Genre')) # R1
movies['Genre'].nunique() # R2
```

```python
# find item(row) in each group

movies.groupby('Genre').size().sort_values(ascending = False) # R1
movies['Genre'].value_counts() # R2
```

```python
# find first and last item from each group

genres = movies.groupby('Genre')
genres.first()
genres.last()
genres.nth(6) # 7th movie of each group (group start with 0)
```

```python
# get_group -> details of particular group

genres.get_group('Horror') # R1
movies[movies['Genre'] == 'Horror'] # R2
```

```python
# which group contains which index item

genres.groups
```

```python
genres.describe()
genres.sample() # 1 random value from each genre
genres.sample(2, replace = True) # 2 random value from each genre if not available repeat first
genres.nunique() # nunique() for -> return count
```

```python
# multiple aggregation (agg method) -> pass dictionary

genres.sum()
genres.agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Metascore': 'min'
    }
)
```

```python
# multiple aggregation -> pass list

genres[['Runtime', 'IMDB_Rating']].agg(['min', 'max', 'mean'])
```

```python
# combined both

genres.agg(
    {
        'Runtime': ['mean', 'max'],
        'IMDB_Rating': 'mean',
        'No_of_Votes': ['sum', 'min'],
        'Metascore': 'min'
    }
)
```

```python
# looping on group

for group, data in genres:
    print(group, data) # dataframe
```

```python
# find the highest rated movie of each 

df = pd.DataFrame(columns = movies.columns)

genres = movies.groupby('Genre')
for group, data in genres:
    filtered_data = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, filtered_data])
df
```

```python
# split - apply 

genres.apply(max, include_groups = False) 
```

```python
# find number of movies starting with A for each group

def func(data):
    return data['Series_Title'].str.startswith('A').sum()

genres.apply(func)
```

```python
# find ranking of each movie in the group according to IMDB rating

def func1(data):
    data['genre_rank'] = data['IMDB_Rating'].rank(ascending = False)
    return data

genres.apply(func1)
```

```python
# find normalized (X - Xmin) / (Xmax - Xmin) IMDB rating group wise

def func2(data):
    data['norm_rank'] = (data['IMDB_Rating'] - data['IMDB_Rating'].min()) / (data['IMDB_Rating'].max() - data['IMDB_Rating'].min())
    return data

genres.apply(func2)
```

```python
# groupby on multiple columns
duo = movies.groupby(['Director', 'Star1'])
# duo.size()

# get_group -> get particular group
duo.get_group(('Aamir Khan', 'Amole Gupte'))
```

```python
# most earning actor-director combo

duo['Gross'].sum().sort_values(ascending = False)
```

```python
# find best(in term of meta score(avg)) actor-genre combo 

movies.groupby(['Star1', 'Genre'])['Metascore'].mean() # this is series
movies.groupby(['Star1', 'Genre'])['Metascore'].mean().reset_index() # this is dataframe
movies.groupby(['Star1', 'Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending = False)
```

```python
# agg on multiple groupby

duo.agg(['min', 'max'])
```

```python
ipl = pd.read_csv('./deliveries.csv')
ipl.shape
```

```python
# find top 10 batsman in terms of runs

ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

```python
# find the batsman with max number of six

# ipl
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).head(10)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs

temp_df = ipl[ipl['over'] > 15]
temp_df1 = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df1.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
# Find V Kohli's record against all team

temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum() # series
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index() # dataframe
```

```python
# create a function that can return the highest score of any batsman

temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

def high_score(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

high_score('V Kohli')
```
