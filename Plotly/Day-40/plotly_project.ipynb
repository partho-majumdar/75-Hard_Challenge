```python
import numpy as np
import pandas as pd
```

```python
latlong = pd.read_csv('./district wise centroids.csv')
latlong.info()
latlong.duplicated().sum()
```

```python
census = pd.read_csv("./india-districts-census-2011.csv")
census.isnull().sum()
census.head()
```

```python
cols = ['District code', 'District name', 'Population', 'Male', 'Female', 'Literate', 'Households_with_Internet', 'Housholds_with_Electric_Lighting']

census = census[cols]
# list(census.columns)
```

```python
final_df = latlong.merge(census, left_on='District', right_on='District name')
```

```python
final_df['sex_ration'] = round((final_df['Female'] / final_df['Male']) * 100)
final_df.sample(5)
```

```python
final_df['literacy_rate'] = round((final_df['Literate'] / final_df['Population']) * 100)
```

```python
final_df.sample(5)
```

```python
final_df = final_df.drop(columns=['Male', 'Female', 'Literate'])
```

```python
final_df.head()
final_df.to_csv('india.csv', index=False)
```
